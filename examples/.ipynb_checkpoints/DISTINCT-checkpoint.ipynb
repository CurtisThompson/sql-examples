# Set-Up

```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect('../data/data.db')
c = conn.cursor()

def execute_statement(statement):
    c.execute(statement)
    res = c.fetchall()
    column_names = [description[0] for description in c.description]
    return pd.DataFrame(res, columns=column_names)
```

# DISTINCT
To get unique values 

```python
execute_statement('SELECT DISTINCT Year FROM cites')
```

# Sorting DISTINCT Results

```python
execute_statement('SELECT DISTINCT Year FROM cites ORDER BY Year ASC')
```
